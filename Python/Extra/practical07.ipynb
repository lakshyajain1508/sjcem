```python
import tkinter as tk
from tkinter import messagebox

# ---- Add Task ----
def add_task():
    task = task_entry.get()
    if task != "":
        listbox.insert(tk.END, task)
        task_entry.delete(0, tk.END)
    else:
        messagebox.showwarning("Input Error", "Please enter a task")

# ---- Delete Task ----
def delete_task():
    try:
        selected = listbox.curselection()[0]
        listbox.delete(selected)
    except:
        messagebox.showwarning("Selection Error", "Please select a task to delete")

# ---- Clear All ----
def clear_all():
    listbox.delete(0, tk.END)

# ---- GUI Setup ----
root = tk.Tk()
root.title("To-Do List")
root.geometry("300x400")

# Entry box
task_entry = tk.Entry(root, width=30)
task_entry.pack(pady=10)

# Buttons
tk.Button(root, text="Add Task", command=add_task).pack(pady=5)
tk.Button(root, text="Delete Task", command=delete_task).pack(pady=5)
tk.Button(root, text="Clear All", command=clear_all).pack(pady=5)

# Listbox
listbox = tk.Listbox(root, width=40, height=15)
listbox.pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

# Function to save student data
def submit_form():
    name = entry_name.get()
    roll = entry_roll.get()
    branch = entry_branch.get()
    email = entry_email.get()

    if name and roll and branch and email:
        messagebox.showinfo("Success", f"Student {name} registered successfully!")
        entry_name.delete(0, tk.END)
        entry_roll.delete(0, tk.END)
        entry_branch.delete(0, tk.END)
        entry_email.delete(0, tk.END)
    else:
        messagebox.showwarning("Input Error", "Please fill all fields!")

# Main window
root = tk.Tk()
root.title("Student Registration Form")
root.geometry("400x400")
root.config(bg="#f0f4ff")

# Heading
title_label = tk.Label(root, text="Student Registration Form", font=("Arial", 16, "bold"), bg="#4a90e2", fg="white", pady=10)
title_label.pack(fill="x")

# Frame for form fields
form_frame = tk.Frame(root, bg="white", bd=2, relief="groove", padx=20, pady=20)
form_frame.place(relx=0.5, rely=0.5, anchor="center")

# Labels and Entries
tk.Label(form_frame, text="Full Name:", font=("Arial", 12), bg="white").grid(row=0, column=0, sticky="w", pady=5)
entry_name = tk.Entry(form_frame, font=("Arial", 12), width=25, bd=2, relief="solid")
entry_name.grid(row=0, column=1, pady=5)

tk.Label(form_frame, text="Roll Number:", font=("Arial", 12), bg="white").grid(row=1, column=0, sticky="w", pady=5)
entry_roll = tk.Entry(form_frame, font=("Arial", 12), width=25, bd=2, relief="solid")
entry_roll.grid(row=1, column=1, pady=5)

tk.Label(form_frame, text="Branch:", font=("Arial", 12), bg="white").grid(row=2, column=0, sticky="w", pady=5)
entry_branch = tk.Entry(form_frame, font=("Arial", 12), width=25, bd=2, relief="solid")
entry_branch.grid(row=2, column=1, pady=5)

tk.Label(form_frame, text="Email ID:", font=("Arial", 12), bg="white").grid(row=3, column=0, sticky="w", pady=5)
entry_email = tk.Entry(form_frame, font=("Arial", 12), width=25, bd=2, relief="solid")
entry_email.grid(row=3, column=1, pady=5)

# Submit button
submit_btn = tk.Button(form_frame, text="Register", font=("Arial", 12, "bold"), bg="#4a90e2", fg="white", padx=10, pady=5, command=submit_form)
submit_btn.grid(row=4, columnspan=2, pady=15)

root.mainloop()
```
